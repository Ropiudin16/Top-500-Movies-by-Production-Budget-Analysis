## 8. Apakah rata - rata profit film bergenre action dan adventure memiliki nilai yang signifikan atau tidak berdasarkan data top 500 Movies by Production Budget?

Uji hipotesis yang digunakan adalah uji t-test dua sampel independent dua arah, dengan hipotesis sebagai berikut:

**Hipotesis**

**H0: μ_genre_action = μ_genre_adventure**

**H1: μ_genre_action != μ_genre_adventure**

```python
New2_df
```

```python
genre_action = New2_df[New2_df['genre']=='Action'][['release_date','profit']].groupby('release_date').sum()
genre_adventure = New2_df[New2_df['genre']=='Adventure'][['release_date','profit']].groupby('release_date').sum()

print('Average profit of genre action a day: ${}'.format(np.round(genre_action.profit.mean())))
print('Average profit of genre adventure a day: ${}'.format(np.round(genre_adventure.profit.mean())))
```

```python
t_stat, p_val = stats.ttest_ind(genre_action,genre_adventure)
print('P-value:',p_val[0])
print('t-statistics:',t_stat[0])
```

```python
genre_action_pop = np.random.normal(genre_action.profit.mean(),genre_action.profit.std(),10000)
genre_adventure_pop = np.random.normal(genre_adventure.profit.mean(),genre_adventure.profit.std(),10000)

ci = stats.norm.interval(0.90, genre_action.profit.mean(), genre_action.profit.std())

fig, ax =plt.subplots(figsize=(16,10))

sns.distplot(genre_action_pop, label='Genre Action Average profit a Day *Pop',kde=True, color='green')
sns.distplot(genre_adventure_pop, label='Genre Adventure Average profil a Day *Pop',kde=True, color='yellow')


ax.axvline(genre_action.profit.mean(), color='green', linewidth=2, label='Action mean')
ax.axvline(genre_adventure.profit.mean(), color='yellow',  linewidth=2, label='Adventure mean')

ax.axvline(ci[1], color='red', linestyle='dashed', linewidth=2, label='confidence threshold of 90%')
ax.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)

ax.axvline(genre_action_pop.mean()+t_stat[0]*genre_action_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
ax.axvline(genre_action_pop.mean()-t_stat[0]*genre_action_pop.std(), color='black', linestyle='dashed', linewidth=2)

ax.margins(0)
plt.title('Grafik Distribusi Genre Action Average profit vs Genre Adventure Average profil')
ax.legend()
plt.savefig('Distribusi.png')
```

> Dari hasil analisa di atas, didapatkan nilai p-value atau nilai probability sampel rata - rata yang diperoleh bukan secara kebetulan yaitu sebesar 0.5712974724301729. Pada code di atas juga saya menggunakan confidence level sebesar 95% atau alpa sebesar 5%

> Berdasarka hasil di atas diketahui bahwa p-value > alpa, yang berarti fail to reject H0 atau gagal menolah H0. Sehingga dapat disimpulan bahwa tidak signifikan antara rata - rata profit film bergenre action dengan rata - rata profit film bergenre adventure
